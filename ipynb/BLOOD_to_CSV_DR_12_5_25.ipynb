```python
Chris_open = r"/Users/chrissoria/Google Drive/Other computers/My Laptop (1)/documents/CADAS/data/CADAS data upload/RD lab results/All_PDF_Blood_Results.csv"
Chris_folder_path = r"/Users/chrissoria/Google Drive/Other computers/My Laptop (1)/documents/CADAS/data/CADAS data upload/RD lab results/All_PDFs"

Ty_open = r"C:\Users\Ty\Desktop\All_PDF_Blood_Results_DR.csv"
#Ty_folder_path = r"G:\.shortcut-targets-by-id\1W2cJLOLxbqzOAwJhLhGSqwAfzOjg0nTo\CADAS data upload\RD lab results\All_PDFs"
Ty_folder_path = r"C:\Users\Ty\Downloads\Berk_Downloads\Clusters 15,13,19,64 68\Cluster 68"

if user == "Ty":
    Chris_open = Ty_open
    Chris_folder_path = Ty_folder_path

folder_path = Chris_folder_path
with open(Chris_open, 'r+') as labels:
    labels.write("SA_ClusterID,SA_HouseID,SA_PersonID,SA_Sex,SA_Age,GlobulosBlancos,GlobulosRojos,Hemoglobina,Hematrocrito,VCM,HCM,CHCM,RDW,Plaquetas,VPM,PDW,Neutrofilos,NeutrofilosP,Linfocitos,LinfocitosP,Monocitos,MonocitosP,Eosinofilos,EosinofilosP,Basofilos,BasofilosP,HbA1c,")
for filename in glob.glob(os.path.join(folder_path, '*.pdf')):
    with open(filename, 'r') as f:
        reader = PdfReader(filename)
        finaltext = "start"
        for i in range(len(reader.pages)):
            page = reader.pages[i]
            finaltext = finaltext + "\n" + page.extract_text()
        if filename == r"G:\.shortcut-targets-by-id\1W2cJLOLxbqzOAwJhLhGSqwAfzOjg0nTo\CADAS data upload\RD lab results\All_PDFs\ZBsKIVt3GLPXaHVqqC4TmLSPJwI59y134323.pdf":
            cluster = "07"
            houseid = "259"
            personid = "1"
        else:
            ID_match = re.search(r'(CLUSTER.{0,15})', finaltext)
            if ID_match:
                ParticID = ID_match.group(1)
                ParticID = ParticID.replace("CLUSTER", "")
                ParticID = ParticID.replace("-", ", ")
                cluster = ParticID.split(",")[0]
                houseid = ParticID.split(",")[1]
                personid = ParticID.split(",")[2]
                personid = personid.strip()
            if personid[0:1] != "0":
                personid = personid[0:1]
            else:
                personid = personid[0:2]
        age_match = re.search(r'(.{0,5}años)', finaltext)
        if age_match:
            age = age_match.group(1)
            age = age.strip()
            age = age.replace("años", "")
            age = age.replace(":", "")
        else:
            age = "."
        sex_match = re.search(r'(Sexo.{0,2})', finaltext)
        if sex_match:
            sex = sex_match.group(1)
            sex = sex.replace("Sexo:","")
        else:
            sex = "."
        globulos_blancos = re.search(r'(Blancos...{0,6})', finaltext)
        if globulos_blancos:
            data1 = globulos_blancos[0]
            data1 = data1.replace("Blancos............K/uL........", "")
            data1 = data1.replace("Blancos............ K/uL........", "")
            data1 = data1.replace("Blancos**", "")
            data1 = data1.replace("*", "")
        else:
            data1 = "."
        globulos_rojos = re.search(r'(Rojos...{0,6})', finaltext)
        if globulos_rojos:
            data2 = globulos_rojos[0]
            data2 = data2.replace("Rojos..............M/uL........", "")
            data2 = data2.replace("Rojos.............. M/uL........", "")
            data2 = data2.replace("Rojos**", "")
            data2 = data2.replace("*", "")
        else:
            data2 = "."
        hemoglobina = re.search(r'(Hemoglobina\*...{0,6})', finaltext)
        if hemoglobina:
            data3 = hemoglobina[0]
            data3 = data3.replace("Hemoglobina.................g/dL........", "")
            data3 = data3.replace("Hemoglobina................. g/dL........", "")
            data3 = data3.replace("Hemoglobina**", "")
            data3 = data3.replace("*", "")
        else:
            data3 = "."
        hematocrito = re.search(r'(Hematocrito[*]?[*]?...{0,6})', finaltext)
        if hematocrito:
            data4 = hematocrito[0]
            data4 = data4.replace("Hematócrito.................%...........", "")
            data4 = data4.replace("Hematócrito................. %...........", "")
            data4 = data4.replace("Hemat�crito................. %...........", "")
            data4 = data4.replace("Hematocrito**", "")
            data4 = data4.replace("*", "")
            data4 = data4.replace("%", "")
        else:
            data4 = "." 
        VCM = re.search(r'(VCM...{0,5})', finaltext)
        if VCM:
            data5 = VCM[0]
            data5 = data5.replace("VCM.........................fL..........", "")
            data5 = data5.replace("VCM", "")
            data5 = data5.replace("*", "")
        else:
            data5 = "."
        HCM = re.search(r'(\nHCM...{0,5})', finaltext)
        if HCM:
            data6 = HCM[0]
            data6 = data6.replace("\nHCM.........................pg..........", "")
            data6 = data6.replace("\nHCM", "")
            data6 = data6.replace("*", "")
        else:
            data6 = "."
        CHCM = re.search(r'(CHCM...{0,5})', finaltext)
        if CHCM:
            data7 = CHCM[0]
            data7 = data7.replace("CHCM........................g/dL........", "")
            data7 = data7.replace("CHCM", "")
            data7 = data7.replace("*", "")
        else:
            data7 = "."
        RDW = re.search(r'(RDW...{0,5})', finaltext)
        if RDW:
            data8 = RDW[0]
            data8 = data8.replace("RDW.........................%...........", "")
            data8 = data8.replace("RDW", "")
            data8 = data8.replace("*", "")
        else:
            data8 = "."
        plaquetas = re.search(r'(Plaquetas[*]?[*]?...{6})', finaltext)
        if plaquetas:
            data9 = plaquetas[0]
            data9 = data9.replace("Plaquetas...................K/uL........", "")
            data9 = data9.replace("Plaquetas**", "")
            data9 = data9.replace("*", "")
        else:
            data9 = "."
        VPM = re.search(r'(MPV...{0,4})', finaltext)        
        if VPM:
            data10 = VPM[0]
            data10 = data10.replace("VPM.........................fL..........", "")
            data10 = data10.replace("MPV", "")
            data10 = data10.replace("*", "")
        else:
            data10 = "."
        PDW = re.search(r'(PDW...{0,5})', finaltext)
        if PDW:
            data11 = PDW[0]
            data11 = data11.replace("PDW.........................GSD.........", "")
            data11 = data11.replace("PDW", "")
        else:
            data11 = "."
        neutrofilos = re.search(r'(Neutrófilos #...{0,4})', finaltext)
        if neutrofilos:
            data12 = neutrofilos[0]
            data12 = data12.replace("Neutrofilos.................K/uL........", "")
            data12 = data12.replace("Neutrófilos #", "")
            data12 = data12.replace("*","")
        else:
            data12 = "."
        neutrofilosP = re.search(r'(Neutrófilos %...{0,5})', finaltext)
        if neutrofilosP:
            data13 = neutrofilosP[0]
            data13 = data13.replace("Neutrófilos %", "")
            data13 = data13.replace("*", "")
        else:
            data13 = "."
        linfocitos = re.search(r'(Linfocitos #...{0,4})', finaltext)
        if linfocitos:
            data14 = linfocitos[0]
            data14 = data14.replace("Linfocitos #", "")
            data14 = data14.replace("Linfocitos %................%...........", "")
        else:
            data14 = "."
        linfocitosP = re.search(r'(Linfocitos %...{0,5})', finaltext)
        if linfocitosP:
            data15 = linfocitosP[0]
            data15 = data15.replace("Linfocitos %................%...........", "")
            data15 = data15.replace("Linfocitos %", "")
        else:
            data15 = "."
        monocitos = re.search(r'(Monocitos #...{0,4})', finaltext)
        if monocitos:
            data16 = monocitos[0]
            data16 = data16.replace("Monocitos #", "")
        else:
            data16 = "."
        monocitosP = re.search(r'(Monocitos %...{0,5})', finaltext)
        if monocitosP:
            data17 = monocitosP[0]
            data17 = data17.replace("Monocitos %", "")
            data17 = data17.replace("*", "")
            data17 = data17.replace("%", "")
        else:
            data17 = "."
        eosinofilos = re.search(r'(Eosinófilos #...{0,4})', finaltext)
        if eosinofilos:
            data18 = eosinofilos[0]
            data18 = data18.replace("Eosinófilos #", "")
        else:
            data18 = "."
        eosinofilosP = re.search(r'(Eosinófilos %...{0,4})', finaltext)
        if eosinofilosP:
            data19 = eosinofilosP[0]
            data19 = data19.replace("Eosinófilos %", "")
        else:
            data19 = "."
        basofilos = re.search(r'(Basófilos #...{0,4})', finaltext)
        if basofilos:
            data20 = basofilos[0]
            data20 = data20.replace("Basófilos #", "")
        else:
            data20 = "."
        basofilosP = re.search(r'(Basófilos %...{0,4})', finaltext)
        if basofilosP:
            data21 = basofilosP[0]
            data21 = data21.replace("Basófilos %", "")
        else:
            data21 = "."
        hba1c = re.search(r'(HPLC...{0,4})', finaltext)
        if hba1c:
            data22 = hba1c[0]
            data22 = data22.replace("HbA1c.......................%...........", "")
            data22 = data22.replace("HPLC", "")
            data22 = data22.replace("*", "")
            data22 = data22.replace("%", "")
        else:
            data22 = "."
        alldata = "\n" + cluster + "," + houseid + "," + personid + "," + sex + "," + age + "," + data1 + "," + data2 + "," + data3 + "," + data4 + "," + data5 + "," + data6 + "," + data7 + "," + data8 + "," + data9 + "," + data10 + "," + data11 + "," + data12 + "," + data13 + "," + data14 + "," + data15 + "," + data16 + "," + data17 + "," + data18 + "," + data19 + "," + data20 + "," + data21 + "," + data22
        print("pdf processed")
        with open(Chris_open, 'a') as out:
            out.write(alldata)
#Remove duplicate lines
clean = pd.read_csv(Chris_open)
clean.drop_duplicates(subset=None, inplace=True)
clean.to_csv(Chris_open, index=False)
print("Finished")
```
