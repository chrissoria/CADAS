```python
# Get the list of all files and directories
dir_list1 = os.listdir(path1)

#to see which files I received from the data manager
newdir_list1 = [x.replace(".epi7", "") for x in dir_list1]

#to see which files I was able to process with epi info tool
dir_list2 = os.listdir(path2)
newdir_list2 = [x.replace(".csv", "") for x in dir_list2]


def remove_common(a, b):
    for i in a[:]:
        if i in b:
            a.remove(i)
            b.remove(i)
 
    print("list1 : ", a)
    print("list2 : ", b)
remove_common(newdir_list1, newdir_list2)

if country == "Cuba":
    numbers_to_filter = ["01_CU", "02_CU", "03_CU", "04_CU", "05_CU", "06_CU", "07_CU", "08_CU", "09_CU", "10_CU", "11_CU"]
elif country == "DR":
    numbers_to_filter = ["01_RD", "02_RD", "03_RD", "04_RD", "05_RD", "06_RD", "07_RD", "08_RD", "09_RD", "10_RD", "11_RD"]

Received_Files = pd.DataFrame(columns=numbers_to_filter)

# Loop through the list and filter list based on numbers above
column_data = {num: [] for num in numbers_to_filter}

for text in newdir_list2:
    for number in numbers_to_filter:
        if number in text:
            column_data[number].append(text)

for number, texts in column_data.items():
    Received_Files[number] = pd.Series(texts)
    
for number, texts in column_data.items():
    Received_Files[number] = pd.Series(texts)

# Drop columns where all values are NA in case a specific tablet's data isn't received
Received_Files.dropna(axis=1, how='all', inplace=True)

# Sort columns alphabetically so that It's easy to see where things are missing
Received_Files = Received_Files.sort_index(axis=1)

Received_Files.head()

Received_Files.to_excel('Received_Files_Checklist.xlsx', index=False)
```
