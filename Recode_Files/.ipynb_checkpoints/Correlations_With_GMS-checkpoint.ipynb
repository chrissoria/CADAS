```python
dfw1 = pd.read_stata('data/1066_Baseline_data.dta', convert_categoricals=False)

dfw1.columns = dfw1.columns.str.lower()
```

```python
## Calculating correlations with f_eurocase
correlations = dfw1.corrwith(dfw1['gmsdiag']).drop('gmsdiag')

# Sorting correlations by absolute values in descending order
sorted_correlations = correlations.abs().sort_values(ascending=False)

# Printing sorted correlations
print("Correlations with gmsdiag, sorted:")
for idx in sorted_correlations.index:
    print(f"{idx}: {correlations[idx]}")

# Finding the variable with the highest absolute correlation
max_corr_var = sorted_correlations.idxmax()
max_corr_value = correlations[max_corr_var]

print("\nThe variable most highly correlated with gmsdiag is", max_corr_var,
      "with a correlation of", max_corr_value)
```
